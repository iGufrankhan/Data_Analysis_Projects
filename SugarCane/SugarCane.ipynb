```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
data_1=pd.read_csv('List of Countries by Sugarcane Production.csv')
```

```python
data_1.head()
```

```python
data_1.isnull()
```

```python
data_1.isnull().sum()
```

## Data Cleaning

```python
data_1.info()
```

```python
data_1['Production (Tons)']=data_1['Production (Tons)'].str.replace(".","")
data_1['Production per Person (Kg)']=data_1['Production per Person (Kg)'].str.replace(".","").str.replace(",",".")
data_1['Acreage (Hectare)']=data_1['Acreage (Hectare)'].str.replace(".","")
data_1['Yield (Kg / Hectare)']=data_1['Yield (Kg / Hectare)'].str.replace(".","").str.replace(",",".")
data_1.head()
```

```python
data_1.isnull().sum()
```

```python
data_1=data_1.dropna().reset_index()

data_1.head()
```

```python
data_1.isnull().sum()
```

```python
data_1 = data_1.drop( "Unnamed: 0", axis = 1)

```

```python
data_1.isnull().sum()
```

```python

data_1 = data_1.drop( "index", axis = 1)
```

```python
data_1.head()
```

```python
data_1.rename(columns= {"Production (Tons)": "Production(Tons)"}, inplace = True)
data_1.rename(columns= {"Production per Person (Kg)": "Production_per_person(Kg)"}, inplace = True)
data_1.rename(columns= {"Acreage (Hectare)": "Acreage(Hectare)"}, inplace = True)
data_1.rename(columns= {"Yield (Kg / Hectare)": "Yield(Kg/Hectare)"}, inplace = True)
```

```python
data_1.head()
```

```python
data_1.info()
```

```python
data_1['Production(Tons)']=data_1['Production(Tons)'].astype(float)
data_1["Production_per_person(Kg)"] = data_1["Production_per_person(Kg)"].astype(float)
data_1["Acreage(Hectare)"] = data_1["Acreage(Hectare)"].astype(float)
data_1["Yield(Kg/Hectare)"] = data_1["Yield(Kg/Hectare)"].astype(float)
data_1.head()
```

## continent analysis

```python
data_1.head()
```

```python
df_continent = data_1.groupby("Continent").sum()
```

```python
df_continent
```

```python
df_continent["number_of_countries"] = data_1.groupby("Continent").count()["Country"]
```

```python
df_continent
```

## Which continent produces maximum sugarcane?

```python
ax=df_continent.sort_values(by="Production(Tons)",ascending=False)
sns.barplot(data=ax,x="Continent",y="Production(Tons)")
```

## Do number of countries in a Continent effects production of sugarcane?

```python
continent_names = df_continent.index.to_list()
sns.lineplot(data = df_continent,x = "number_of_countries", y= "Production(Tons)" )
plt.xticks(df_continent["number_of_countries"], continent_names, rotation =90)
plt.show()
```

## Do continent with highest land produces more sugarcane?

```python
sns.lineplot(data = df_continent,x = "Acreage(Hectare)", y= "Production(Tons)" )
```
